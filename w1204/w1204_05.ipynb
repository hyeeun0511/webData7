```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import requests
from bs4 import BeautifulSoup
import time  
import os
import random
```

```python
url = "https://nid.naver.com/nidlogin.login?mode=form&url=https://www.naver.com/"
browser = webdriver.Chrome()

# 1. naver 페이지 열기
browser.get(url)

time.sleep(2)

# input데이터 넣기
input_js = '''
document.getElementById("id").value = "{id}";
document.getElementById("pw").value = "{pw}";
'''.format(id='bang48188',pw='grasia48188')

time.sleep(2)  # 3초 대기
# time.sleep(random.uniform(1,5))  # 1~5초 사이에 랜덤으로 돌려서 대기

# naver에서 elem라는 요소로 찾기해서 데이터 입력시 차단
# 자바스크립트를 사용해서 데이터 입력
browser.execute_script(input_js)  # 스크립트 실행  # input됨
time.sleep(2)
browser.find_element(By.ID,"log.login").click()
```

```python
# 메일 버튼 클릭
browser.find_element(By.CLASS_NAME,"MyView-module__item_text___VTQQM").click()
```

```python
# 메일 쓰기 버튼 클릭
browser.find_element(By.XPATH,'//*[@id="account"]/div[3]/div[2]/div[2]/a').click()
```

```python
# 탭 이동
# 새 탭이기 때문에 탭 이동 후 메일 주소 입력가능
browser.switch_to.window(browser.window_handles[1])
```

```python
time.sleep(2)
# 보낼 메일 주소 입력(나에게 보내기)
elem = browser.find_element(By.ID,"recipient_input_element")
elem.send_keys("bang48188@naver.com")
```
